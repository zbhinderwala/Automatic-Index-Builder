```python
import pandas as pd
import numpy as np
```

```python
index_properties = pd.read_csv("Properties.csv", low_memory=False)
```

```python
doc_count = 3332956
index_properties = index_properties.fillna(1)
#index_properties = index_properties['volume_count'].apply(pd.to_numeric)
index_properties['volume_count'] = pd.to_numeric(index_properties.volume_count, errors='coerce')
x =  index_properties['frequency']/index_properties['wordCount'] 
y =  index_properties['volume_count'].apply(lambda x: x/doc_count)
index_properties['tf_idf'] = x / y                                                   
```

```python
index_properties['pos_rank'] = 2
index_properties.loc[index_properties.pos == 'NN' , 'pos_rank'] = 14
index_properties.loc[index_properties.pos == 'JJ' , 'pos_rank' ] = 7
index_properties.loc[index_properties.pos == 'NNS' , 'pos_rank'] = 6
index_properties.loc[index_properties.pos == 'NNP' , 'pos_rank' ] = 6
index_properties.loc[index_properties.pos == 'VBP' , 'pos_rank' ] = 3
```

```python
index_properties['score'] = 0
for index,row in index_properties.iterrows():
    score = ((row['pos_rank'] * row['n_gram_score']))+row['tf_idf'] 
    index_properties.loc[index , 'score'] = score
```

```python
def match_count(predicted, actual):
    count = 0
    pred_copy = predicted
    act_copy = actual
    for str in predicted:
        if (actual.count(str) >= 1):
            count = count + 1
            act_copy.remove(str)
            pred_copy.remove(str)
    left_pred = pred_copy
    left_act = act_copy[:]
    match = False
    for index in act_copy:
            for str in pred_copy:
                if(index.find(str) != -1):
                    count = count + 1
                    match = True
                    left_pred.remove(str)
                    
            if match:
                left_act.remove(index)
    list = [count, len(pred_copy),len(act_copy)]
    return list

        
```

```python
res_idx_1 = match_count(list(index_properties_1['word'][:23].values), list(indexes_1['word'].values))
```

```python
def find_precision_recall(result):
    tp = result[0]
    fp = result[1]
    fn = result[2]
    
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    
    evaluation = [precision,recall]
    return evaluation
```

```python
eval1 = find_precision_recall(res_idx_1)
eval2 = find_precision_recall(res_idx_2)
eval3 = find_precision_recall(res_idx_3)
eval4 = find_precision_recall(res_idx_4)
print( eval1 , eval2 , eval3 , eval4)
```

```python
avg_precision = (eval1[0]+eval2[0]+eval3[0]+eval4[0])/4
avg_recall = (eval1[1]+eval2[1]+eval3[1]+eval4[1])/4
print( avg_precision, avg_recall)
```

```python
def match_count1(predicted, actual):
    count = 0
    index_words = []
    pred_copy = predicted
    act_copy = actual
    for str in predicted:
        if (actual.count(str) >= 1):
            count = count + 1
            act_copy.remove(str)
            pred_copy.remove(str)
            index_words.append(str)
    left_pred = pred_copy
    left_act = act_copy[:]
    match = False
    for index in act_copy:
            for str in pred_copy:
                if(index.find(str) != -1):
                    count = count + 1
                    match = True
                    left_pred.remove(str)
                    index_words.append(str)
            if match:
                left_act.remove(index)
    list = [count, len(pred_copy),len(act_copy),index_words]
    return list
```
